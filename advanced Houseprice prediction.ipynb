### HANDLING CATEGORICAL MISSING VALUES

```python
categorical_engg = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype=="O"]
```

```python
categorical_engg
```

```python
for feature in categorical_engg:
    data = df.copy()
    df[feature]=df[feature].fillna('Missing')
```

```python
df[categorical_engg].isnull().sum()
```

### HANDLING NUMERICAL  MISISNG VALUES

```python
numerical_engg = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype!="O" ]
```

```python
numerical_engg
```

```python

   
df[numerical_engg].isnull().sum()    
```

```python
for feature in numerical_engg:
    median_value = df[feature].median()
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature]=df[feature].fillna(median_value)
```

```python
df[numerical_engg].isnull().sum()
```

### HANDLING TEMPORAL VALUES OF NUMERICAL FEATURE

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

### APPLYING LOG NORMAL DISTRIBUTION FOR CATEGO VALUES FOR FEATURE WITHOUT ZEROS

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

### HANDLING RARE CATEGORICAL VALUES

```python
for feature in categorical_values:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
    
```

```python
df.head(20)
```

### LABELING CATEGORICAL VALUES WITH NUMERICAL VALUES BY ENUMERATE

```python
for feature in categorical_values:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

## Feature selection

```python
X=dataa.drop(['Id','SalePrice'],axis=1)
```

```python
X.isnull().any()
```

```python
X.head()
```

```python
y=dataa[['SalePrice']]
```

```python
y.head()
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model =SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X,y)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat = X.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```
